### Gaussian Error Linear Unit (GELU)

```python
import numpy as np
```

image.png

image.png

```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))
```

```python
x = np.linspace(-10, 10)
x
```

```python
import matplotlib.pyplot as plt
```

```python
y = gelu(x)
print(y)
plt.plot(x,y)
plt.title("GELU")
```

```python
# smooth and differential everywhere unlike ReLU, helps in gradient flow,
# also activation is on how much input exceeds other inputs,
# for weights by their value rather than input's sign.
```
