```python
import numpy as np
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def leakyrelu(x):
    return np.maximum(0.01 * x, x)
```

```python
def prelu(x, alpha):
    return np.maximum(alpha * x, x)
```

```python
def elu(x, alpha):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
x = np.linspace(-10, 10) #evenly spaced values 
x
```

```python
import matplotlib.pyplot as plt
```

```python
y = relu(x)
print(y)
plt.plot(x,y)
plt.title("ReLU")
```

```python
y = leakyrelu(x)
print(y)
plt.plot(x,y)
plt.title("Leaky ReLU")
```

```python
alpha = 0.3
y = prelu(x, alpha)
print(y)
plt.plot(x,y)
plt.title("PReLU")
```

```python
alpha = 0.4
y = elu(x, alpha)
print(y)
plt.plot(x,y)
plt.title("eLU")
```
